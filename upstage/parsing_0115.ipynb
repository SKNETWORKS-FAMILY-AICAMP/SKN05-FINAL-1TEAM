```python
def json_data_load(file_path):
    with open(file_path,"r",encoding="utf-8") as file:
        return json.load(file)


def split_data_merge(merge_data, plus_data_path, split_num):
    plus_data = json_data_load(plus_data_path)


    merge_data['content']['html'] += "\n"
    merge_data['content']['html'] += plus_data['content']['html']

    merge_data['content']['markdown'] += "\n"
    merge_data['content']['markdown'] += plus_data['content']['markdown']


    for i in plus_data['elements']:
        i['page'] += (split_num-1)*100

    merge_data['elements'].extend(plus_data['elements'])
    
    return merge_data
```

```python
import fitz  # PyMuPDF
import os
from tqdm import tqdm

# PDF íŒŒì¼ ê²½ë¡œì™€ ì¶œë ¥ í´ë” ì„¤ì •

pdf_path = file[0]  # PDF íŒŒì¼ ê²½ë¡œ
cam_name = "a6400"
output_folder = os.path.abspath(f"parsing/pdf_image/{cam_name}")  # ì €ì¥í•  í´ë” ê²½ë¡œ

# ì¶œë ¥ í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# PDF ì—´ê¸°
pdf_document = fitz.open(pdf_path)

# í•´ìƒë„ ì„¤ì • (DPI)
dpi = 300  # ì›í•˜ëŠ” DPI ê°’ (300, 600 ë“±)

# ê° í˜ì´ì§€ë¥¼ ì´ë¯¸ì§€ë¡œ ì €ì¥
for page_number in tqdm(range(len(pdf_document))):
    page = pdf_document.load_page(page_number)  # í˜ì´ì§€ ë¡œë“œ
    pix = page.get_pixmap(dpi=dpi)  # í˜ì´ì§€ë¥¼ ì´ë¯¸ì§€ë¡œ ë Œë”ë§
    output_file = os.path.join(output_folder, f"{cam_name}_page_{page_number + 1}.jpg")
    pix.save(output_file)  # ì´ë¯¸ì§€ ì €ì¥

pdf_document.close()
```

```python
## ì´ë¯¸ì§€ ì¶”ì¶œ ë° ì €ì¥

from PIL import Image
import json

# ì´ë¯¸ì§€ íŒŒì¼ ê²½ë¡œ
# image_path = glob.glob(f'{os.path.abspath("")}/images/*.jpg')
cam_name = 'a6400'
image_dir = os.path.abspath(f"parsing/pdf_image/{cam_name}")
save_image_dir = os.path.abspath(f"parsing/pdf_cropimage/{cam_name}")
path_dir = os.path.abspath("parsing/parsing_result_last")
file_path = f"{path_dir}/{cam_name}_upstage_merge.json"

# ì¶œë ¥ í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±
if not os.path.exists(save_image_dir):
    os.makedirs(save_image_dir)

# JSON íŒŒì¼ ì½ê¸°
with open(file_path, "r", encoding="utf-8") as file:
    response = json.load(file)

filtered_elements = [
    element for element in response['elements']
    if element.get('category') == 'figure'
]

# ìƒˆë¡œìš´ ë”•ì…”ë„ˆë¦¬ì— ì €ì¥
filtered_response = {'elements': filtered_elements}


# ì´ë¯¸ì§€ ì—´ê¸°
page_dump = 0

for coordinate in tqdm(filtered_response['elements']):
    coordinates = coordinate['coordinates']
    page_num = coordinate['page']

    try:
        with Image.open(f"{image_dir}/{cam_name}_page_{page_num}.jpg") as img:
            if page_num != page_dump:
                page_index = 1
            else:
                page_index += 1

            width, height = img.size

            # ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œ ê³„ì‚°
            x_min = int(min(coord['x'] * width for coord in coordinates))
            y_min = int(min(coord['y'] * height for coord in coordinates))
            x_max = int(max(coord['x'] * width for coord in coordinates))
            y_max = int(max(coord['y'] * height for coord in coordinates))

            # ë°”ìš´ë”© ë°•ìŠ¤ í¬ë¡­
            cropped_image = img.crop((x_min, y_min, x_max, y_max))

            # ê²°ê³¼ ì´ë¯¸ì§€ ì €ì¥
            output_path = f'{save_image_dir}/{cam_name}_page_{page_num}_{page_index}.png'
            cropped_image.save(output_path)
            page_dump = page_num

    except FileNotFoundError:
        print(f"File not found: {image_dir}/{cam_name}_page_{page_num}.png")
```
