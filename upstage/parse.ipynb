```python
# bring in our LLAMA_CLOUD_API_KEY
import os
from dotenv import load_dotenv
load_dotenv()
LLAMA_CLOUD_API_KEY = os.getenv("API_KEY")
# bring in deps
from llama_parse import LlamaParse
from llama_index.core import SimpleDirectoryReader

# set up parser
parser = LlamaParse(
    result_type="markdown"  # "markdown" and "text" are available
)

# use SimpleDirectoryReader to parse our file
file_extractor = {".pdf": parser}
documents = SimpleDirectoryReader(input_files=['./data/test/sony_a7m3.pdf'], file_extractor=file_extractor).load_data()
print(documents)
```

```python
import nest_asyncio
import asyncio
from dotenv import load_dotenv
from llama_parse import LlamaParse
from llama_index.core import SimpleDirectoryReader

# 1. Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()

# 2. Load environment variables (API key, etc.)
load_dotenv()

# 3. Set up the parser
parser = LlamaParse(
    result_type="markdown"  # "markdown" and "text" are available
)

# 4. Use SimpleDirectoryReader to parse our file asynchronously
async def parse_documents():
    file_extractor = {".pdf": parser}
    documents = await SimpleDirectoryReader(input_files=['./data/test/sony_a7m3.pdf'], file_extractor=file_extractor).aload_data()
    print(documents)

# 5. Run the async function using asyncio
asyncio.run(parse_documents())
 
```

```python
# one extra dep
from llama_index.core import VectorStoreIndex

# create an index from the parsed markdown  
index = VectorStoreIndex.from_documents(documents)

# create a query engine for the index
query_engine = index.as_query_engine()

# query the engine
query = "카메라"
response = query_engine.query(query)
print(response)
```
